### Ejercicio 1
Considera la siguiente función

\begin{equation}
f(y) = \frac{1}{k(2\pi)^{\frac{3}{2}}} \exp\left[-\frac{1}{172}[31y_1^2+14y_2^2+21y_3^2-4y_1y_2+14y_1y_3-12y_2y_3]\right].
\label{eqejer1}
\end{equation}

**i.** - Muestra que la función anterior es la densidad conjunta de $Y = (Y_1, Y_2, Y_3)^t \sim \mathcal{N}_3(\mu, \Sigma)$ obteniendo $k, \mu, \Sigma$ y $\Sigma^{-1}$ para que se satisfaga la Def. 1.N vista en clase.

Según la definición 1.N, la densidad normal conjunta para un vector $Y=(Y_1,\cdots,Y_n)$ se define como

\begin{equation}
f(y) = \frac{1}{(2\pi)^{\frac{n}{2}}\vert\Sigma\vert^{\frac{1}{2}}} \exp\left[ -\frac{1}{2} (y - \mu)^t \Sigma^{-1} (y - \mu) \right],
\label{eqdefinormal}
\end{equation}

donde $\mu$ y la matriz positiva definida (p. d.) $\Sigma$ representan respectivamente al vector medio y a la matriz de varianzas y covarianzas del vector aleatorio $Y$. \footnote{Esto último no forma parte de la definición 1.N, sino que es un resultado presentado en las notas de clase.}

Notemos primero que, por propiedades básicas de matrices:

$$
\begin{split}
  (y - \mu)^t \Sigma^{-1} (y - \mu) &= \left( (y - \mu)^t \Sigma^{-1}\right) y - \left( (y - \mu)^t \Sigma^{-1}\right) \mu\\
  &= \left( y^t \Sigma^{-1} - \mu^t \Sigma^{-1} \right) y - \left( y^t \Sigma^{-1} - \mu^t \Sigma^{-1}\right) \mu\\
  &= y^t \Sigma^{-1} y - \mu^t \Sigma^{-1} y - y^t \Sigma^{-1} \mu + \mu^t \Sigma^{-1} \mu.\\
\end{split}
$$

Notemos que el término $\mu^t \Sigma^{-1} \mu$ corresponde a sumandos independientes del vector $y$, es decir, sólo depende de $\mu$ y de $\Sigma^{-1}$. Por otra parte, al ser $\Sigma$ una matriz p. d. y simétrica, entonces su inversa $\Sigma^{-1}$ también es p. d. y simétrica. \footnote{Como $\Sigma$ es simétrica, entonces $\Sigma=\Sigma^t$ esto implica que $\Sigma^{-1}=(\Sigma)^{-1}=(\Sigma^t)^{-1}=(\Sigma^{-1})^t$. Por tanto, $\Sigma^{-1}$ es simétrica.} Lo anterior implica que el término $\mu^t \Sigma^{-1} \mu$ es cero sí y sólo si el vector $\mu$ es el vector nulo. Notemos que el argumento de la exponencial en la función \ref{eqejer1} sólo tiene términos que contienen alguna componente del vector $y$, esto implica que si dicha función es la densidad conjunta de una normal multivariada (en el sentido de la definición 1.N) entonces dicha densidad necesariamente tendrá a $\mu=\underline{0}$ como parámetro.

Por tanto, la densidad \ref{eqdefinormal} en 1.N se reduce (en nuestro caso) a

\begin{equation}
f(y) = \frac{1}{(2\pi)^{\frac{3}{2}}\vert\Sigma\vert^{\frac{1}{2}}} \exp\left[ -\frac{1}{2} y^t \Sigma^{-1} y \right].
\label{eqdefinormal2}
\end{equation}


Por lo anterior, para concluir que la función \ref{eqejer1} es la densidad conjunta de $Y = (Y_1, Y_2, Y_3)^t \sim \mathcal{N}_3(\underline{0}, \Sigma)$ es suficiente encontrar la matriz de de varianzas y covarianzas $\Sigma$, tal que $k=\vert\Sigma\vert^\frac{1}{2}$ y $\frac{1}{86}[31y_1^2 + 14 y_2^2+21y_3^2-4y_1y_2+14y_1y_3-12y_2y_3=y^t \Sigma^{-1} y$

Es conveniente desarrollar $y^t \Sigma^{-1} y$ para facilitar la comparación:

$$
\begin{split}
y^t \Sigma^{-1} y &= (y_1,y_2,y_3)
\begin{bmatrix}
c_{11} & c_{12} & c_{13} \\
c_{12} & c_{22} & c_{23} \\
c_{13} & c_{23} & c_{33} \\
\end{bmatrix}
\begin{pmatrix}
y_1 \\
y_2 \\
y_3
\end{pmatrix}\\
&= c_{11}y_1^2 + c_{22}y_2^2 + c_{33}y_3^2 + 2 c_{12}y_1y_2 + 2 c_{13}y_1y_3 + 2 c_{23}y_2y_3
\end{split}
$$

Mediante una comparación directa, se obtiene:

```r
# Función para convertir la matriz a formato LaTeX con bmatrix e imprimir con texto adicional (de manera opcional se puede añadir texto antes y después de la matriz)
matriz_a_latex <- function(mat, pre_text = "", post_text = "") {
  entries <- apply(mat, 1, function(row) paste(row, collapse = " & "))
  body <- paste(entries, collapse = " \\\\\n")
  latex_mat <- paste("\\begin{bmatrix}", body, "\\end{bmatrix}", sep = "\n")

  cat("$$\n", pre_text, latex_mat, post_text, "\n$$", sep = " ")
}

# Función que calcula la matriz adjunta.
calcular_matriz_adjunta <- function(A) {
  n <- nrow(A)
  adjunta <- matrix(0, n, n)

  for (i in 1:n) {
    for (j in 1:n) {
      # Extraer la submatriz excluyendo la fila i y columna j
      submatriz <- A[-i, -j, drop = FALSE]  # Asegura que el resultado sea una matriz

      # Calcular el determinante de la submatriz
      if (is.matrix(submatriz)) {
        menor <- det(submatriz)
      } else {
        # Si submatriz es un vector (caso de matrices 2x2), trátalo como una matriz 1x1
        menor <- submatriz
      }

      # Aplicar el tablero de ajedrez de signos y asignar a la matriz adjunta
      adjunta[i, j] <- (-1)^(i + j) * menor
    }
  }

  # La matriz adjunta es la transpuesta de la matriz de cofactores
  return(t(adjunta))
}

# Definir la matriz
C <- matrix(c(31, -2, 7,
                  -2, 14, -6,
                  7, -6, 21), byrow = TRUE, nrow = 3)

C_ad <- calcular_matriz_adjunta(C)
det_C <- det(C)
sigma <- 86*C_ad/det_C
k=sqrt(det(sigma))
```

```r
#results='asis'
matriz_a_latex(C,"\\Sigma^{-1} = \\frac{1}{86}",".")
```

La matriz adjunta (la matriz de los menores) de $\Sigma^{-1}$ es:



```r
#results='asis'
matriz_a_latex(C_ad,"(\\Sigma^{-1})^+ = \\frac{1}{86^2}",".")
```

Por otra parte, el determinante de $\Sigma^{-1}$ es `r det_C` $/86^3=$ `r det_C/ 86^3`.

Como la matriz adjunta de $\Sigma^{-1}$ es simétrica, entonces la matriz inversa de $\Sigma^{-1}$ se obtiene mediante:




```r
#results='asis'
texto_previo <- paste("\\Sigma = \\frac{((\\Sigma^{-1})^+)^t}{\\vert\\Sigma^{-1}\\vert} = \\frac{(\\Sigma^{-1})^+}{\\vert\\Sigma^{-1}\\vert} = \\frac{",86,"}{", round(det_C,0), "}", sep = "")
matriz_a_latex(C_ad, texto_previo, ".")
```

Si efectuamos los cálculos, la matriz de varianzas y covarianzas quedaría:



```r
#results='asis'
matriz_a_latex(round(sigma,0),"\\Sigma = ",".")
```

Luego, la función \ref{eqejer1} es una densidad normal multivariada con los parámetros encontrados, si la constante $k=\vert\Sigma\vert^\frac{1}{2}=$ `r k`.
